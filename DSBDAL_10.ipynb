```python
import pandas as pd 
from sklearn.datasets import load_iris 
```

```python

# Load the Iris dataset 
iris = load_iris() 

# Convert the dataset into a pandas DataFrame 
data = pd.DataFrame(iris.data, columns=iris.feature_names) 
```

```python

# Add the target variable 'species' to the DataFrame 
data['species'] = iris.target_names[iris.target] 

# Display the first few rows to check the dataset 
data.head()
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python

# Create a histogram for each numeric feature in the dataset 
numeric_features = data.drop(columns='species') 

plt.figure(figsize=(15, 10)) 

# Loop over each numeric feature and plot its histogram 
for i, feature in enumerate(numeric_features.columns, 1): 
    plt.subplot(2, 2, i) 
    sns.histplot(data[feature], kde=True, bins=15) 
    plt.title(f'Distribution of {feature}') 

plt.tight_layout() 
plt.show()
```

```python
plt.figure(figsize=(15, 10)) 

# Loop over each numeric feature and plot its box plot 
for i, feature in enumerate(numeric_features.columns, 1): 
    plt.subplot(2, 2, i) 
    sns.boxplot(x='species', y=feature, data=data) 
    plt.title(f'Box plot of {feature}') 

plt.tight_layout() 
plt.show()
```

```python
# Create box plots for each numeric feature 
plt.figure(figsize=(12, 6)) 

for i, column in enumerate(data.select_dtypes(include='number').columns): 
    plt.subplot(1, 4, i + 1) 
    sns.boxplot(y=data[column], color='lightgreen') 
    plt.title(column) 

plt.suptitle("Box Plots of Numeric Features - Iris Dataset", fontsize=14) 
plt.tight_layout() 
plt.show()
```


## Steps of the Analysis

1. **Load the Iris Dataset:**
The Iris dataset was successfully loaded from the `sklearn.datasets` library. It contains four features: sepal length, sepal width, petal length, and petal width, along with the species labels.
2. **Convert to DataFrame:**
The dataset was converted into a pandas DataFrame, making it easier to manipulate and analyze. The DataFrame includes the four features and the species column.
3. **Display Sample Records:**
The first five records of the dataset were displayed. These records provided an overview of the data and the corresponding species for each flower.
4. **Visualize Distribution (Histogram):**
Histograms were plotted for the four features: sepal length, sepal width, petal length, and petal width. 

5. **Visualize Distribution by Species (Boxplot):**
Boxplots were plotted to visualize the distribution of the sepal and petal dimensions for each species. 

6. **Overall Distribution (Boxplot):**
Boxplots for the overall distribution of the sepal and petal lengths and widths.



## Observations:
1. **Histograms:** Show the distribution of sepal length, sepal width, petal length, and petal width. Some features appear to be normally distributed, while others are skewed.

2. **Box Plots (by species):** Reveal differences in feature distributions across species. For example, setosa generally has smaller petal length and width compared to versicolor and virginica.

3. **Consolidated Box Plots:** Give a compact overview of the range and variability of each numeric feature.
